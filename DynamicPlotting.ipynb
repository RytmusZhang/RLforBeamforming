```python
"Dynamic plotting in matplotlib. Copy and paste into a Jupyter notebook."
# written October 2016 by Sam Greydanus
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import time
from scipy.io import loadmat

FileName='GLOBAL_RUNNING_R_NT2_A3CPolicyWithoutCache_AddingLayer200_2000steps'
fig,ax = plt.subplots(1,1)
ax.set_xlim(0,2000)
ax.set_ylim(-3.5e8,0)
X, Y = [], []
while True:
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    R = loadmat(FileName+'.mat')
    Y = R[FileName]
    X = np.arange(len(Y[0]))
    ax_last = ax.plot(X, Y[0])
    fig.canvas.draw()
    time.sleep(5)
```
